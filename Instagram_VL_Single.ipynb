## 3️⃣ VL Processing

```python
def download_image(url):
    if not url: return None
    try:
        r = requests.get(url,
            headers={'User-Agent': 'Mozilla/5.0', 'Referer': 'https://www.instagram.com/'},
            timeout=15)
        if r.status_code == 200 and len(r.content) > 1000:
            return r.content
    except: pass
    return None

def analyze(img_bytes, caption):
    if not img_bytes: return None
    try:
        model = genai.GenerativeModel('gemini-1.5-flash')
        img = Image.open(io.BytesIO(img_bytes))
        prompt = (
            'Return ONLY valid JSON (no markdown): '
            '{"vlSubject": "subject (5-15 words)", '
            '"mood": "Vibrant|Calm|Minimalist|Nostalgic|Energetic", '
            '"visual_tags": ["t1","t2","t3","t4","t5"], '
            '"narrative": "1-2 sentence description", '
            '"aestheticScore": 1-10}}\n'
            f'Caption: {caption[:300] if caption else "No caption"}'
        )
        resp = model.generate_content([prompt, img])
        text = resp.text.strip()
        if '{' in text and '}' in text:
            result = json.loads(text[text.find('{'):text.rfind('}')+1])
            if 'vlSubject' in result:
                result['provider'] = 'gemini'
                return result
    except: pass
    return None

print('✅ Functions ready')
```
